## Processing for the X y split
## X=features y=target


```python
X=data_red.drop('ova', axis=1)     
y=data_red['ova']
```

```python
# drop name column

X=X.drop(columns=['name'],axis=1)
```

### Again separating numerical and categorical for X-features

```python
X_num1 = X.select_dtypes(np.number)
X_cat1 = X.select_dtypes(object)
```

```python
X_num1.shape
```

```python
X_num1.describe()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

### Scaling using MinMax

```python
# good for counts or if all data are positive numbers
transformer = MinMaxScaler().fit(X_num1)
X_norm = transformer.transform(X_num1)
print(X_norm.shape)
X_num1_scale = pd.DataFrame(X_norm, columns=X_num1.columns)
X_num1_scale.head()
```

## One hot encode the categorical data

```python
# import the library
from sklearn.preprocessing import OneHotEncoder
```

```python
pip install -U scikit-learn
```

```python
# Create encoder to be used on new data later.

encoder = OneHotEncoder(drop='first').fit(X_cat1)
cols = encoder.get_feature_names_out(input_features=X_cat1.columns)
X_cat1_encode = pd.DataFrame(encoder.transform(X_cat1).toarray(),columns=cols)
X_cat1_encode.head()
```

### concat numerical and categorical transformations for features

```python
X=pd.concat([X_num1_scale, X_cat1_encode], axis=1) 
X.describe()
```

### Train Test Split

```python
# import library
from sklearn.model_selection import train_test_split as tts

X_train, X_test, y_train, y_test=tts(X, y, test_size=.2)
```

```python
X_train
```

```python
y_train
```

## Train and run model

```python
#import library
from sklearn.linear_model import LinearRegression as linreg

lm = linreg()    # model
lm.fit(X_train, y_train)   # model training
y_pred = lm.predict(X_test)   # model prediction
```

### Validation R2 scores

```python
print ('train R2: {} -- test R2: {}'.format(lm.score(X_train, y_train),
                                            lm.score(X_test, y_test)))
```

### Mean Squared Error

```python
from sklearn.metrics import mean_squared_error as mse


train_mse=mse(lm.predict(X_train), y_train)
test_mse=mse(y_pred, y_test)

print ('train MSE: {} -- test MSE: {}'.format(train_mse, test_mse))
```

### Mean Absolute Error

```python
from sklearn.metrics import mean_absolute_error as mae

train_mae=mae(lm.predict(X_train), y_train)
test_mae=mae(y_pred, y_test)

print ('train MAE: {} -- test MAE: {}'.format(train_mae, test_mae))
```
